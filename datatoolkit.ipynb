4. Using the given x and y data, generate a 3D surface plot to visualize the function

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)
fig=px.scatter_3d(df,x='X',y='Y',z='Z')
fig.show()
```

BOKEH ASSIGNMENT:

1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
from bokeh.plotting import figure , output_file , show
import numpy as np
x=np.linspace(0,10,100)
y=np.sin(x)
output_file("sine_wave.html")
p = figure(title="Sine Wave", x_axis_label="x", y_axis_label="sin(x)")
p.line(x, y, line_width=2)
show(p)
```

2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook


N = 100
x = np.random.rand(N)
y = np.random.rand(N)
sizes = np.random.randint(5, 20, size=N)
colors = np.random.choice(['red', 'blue', 'green', 'orange'], size=N)
data = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})
output_notebook()
p = figure(title="Scatter Plot with Random Data", x_axis_label="x", y_axis_label="y")
p.scatter(x='x', y='y', size='sizes', color='colors', source=data)
show(p)
```

3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook


fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
output_notebook()
p = figure(x_range=fruits, title="Fruit Counts", toolbar_location=None, tools="")
p.vbar(x=fruits, top=counts, width=0.9)
p.xgrid.grid_line_color = None
p.y_range.start = 0
show(p)
```

4. Create a Bokeh histogram to visualize the distribution of the given data.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook


data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)
output_notebook()
p = figure(title="Histogram of data_hist", tools="", background_fill_color="#fafafa")
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color="navy", line_color="white", alpha=0.5)
p.y_range.start = 0
p.xaxis.axis_label = 'Value'
p.yaxis.axis_label = 'Frequency'

show(p)
```

5. Create a Bokeh heatmap using the provided dataset.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import LinearColorMapper, ColorBar
from bokeh.io import output_notebook


data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)


output_notebook()

data_heatmap_1d = data_heatmap.flatten()


color_mapper = LinearColorMapper(palette="Viridis256", low=data_heatmap_1d.min(), high=data_heatmap_1d.max())


p = figure(title="Heatmap", x_range=(x.min(), x.max()), y_range=(y.min(), y.max()),
           toolbar_location=None, tools="", x_axis_location="above")


p.rect(x=xx.flatten(), y=yy.flatten(), width=x[1] - x[0], height=y[1] - y[0],
       fill_color={'field': 'value', 'transform': color_mapper},
       line_color=None, source={'data': {'value': data_heatmap_1d}})


color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0))
p.add_layout(color_bar, 'right')


show(p)
```
